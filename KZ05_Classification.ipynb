### Visualize A Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from IPython.display import Image
from sklearn import tree
import pydotplus
import graphviz
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel(r"C:\Users\User\Dropbox\sys_eng\Hypothesis(ML)\Data\ForDataModelling.xlsx",engine="openpyxl")
print(data.shape)
data.head()
```

```python
X = data.drop(['Category', 'SKU', 'Freq','Price','Net Sale'], axis = 1)
y = data['Category']
print(X.shape,"\n")
print(y.shape)
```

```python
for column in X.columns:
    if X[column].dtype == np.object:
        dum_col = pd.get_dummies(X[column], prefix = column, drop_first = True)
        X = pd.concat([X, dum_col], axis = 1).drop([column], axis = 1)
print("Shape: ",X.shape)
print("All Columns: ",X.columns)
```

```python
y = pd.get_dummies(y, prefix = 'category', drop_first = True)
print(y.shape)
print(y.columns)
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
clf = DecisionTreeClassifier(random_state = 0)
model = clf.fit(X_train,y_train)
```

```python
X_train.columns
```

```python
dot_data = tree.export_graphviz(clf, out_file = None, 
                                feature_names = X_train.columns,
                                class_names = ['category_Beauty and Health','category_Toys','category_Footwear','category_Accessories','category_Apparel','category_Sunglasses','category_Lifestyle','category_FMCG'])
graph = pydotplus.graph_from_dot_data(dot_data)

Image(graph.create_png())
```
