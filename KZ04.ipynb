```python
data = pd.read_excel(r"C:\Users\User\Dropbox\sys_eng\Hypothesis(ML)\Data\ForDataModelling.xlsx",engine="openpyxl")
print(data.shape)
data.head()
```

```python
X = data.drop(['Category', 'SKU', 'Freq','Month','Month_Num','Year','Price'], axis = 1)
y= data['Category']
print(X.shape,"\n")
print(y.shape)
```

```python
X.columns
```

```python
X = X.drop(['Net Sale'], axis = 1)
```

```python
for column in X.columns:
    if X[column].dtype == np.object:
        dum_col = pd.get_dummies(X[column], prefix = column, drop_first = True)
        X = pd.concat([X, dum_col], axis = 1).drop([column], axis = 1)
print("Shape: ",X.shape)
print("All Columns: ",X.columns)
```

```python
y = pd.get_dummies(y, prefix = 'category', drop_first = True)
print(y.shape)
print(y.columns)
y.head()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

rf = RandomForestClassifier(n_estimators = 20)
rf.fit(X_train, y_train)

prediction = rf.predict(X_test)
prediction_proba = rf.predict_proba(X_test)

print("Accuracy: ", accuracy_score(y_test, prediction))
#print("AUC: ", roc_auc_score(y_test, prediction_proba[:,1]))
```

```python
print(rf)
```

```python
rf2 = RandomForestClassifier(n_estimators = 100)
rf2.fit(X_train, y_train)
prediction = rf2.predict(X_test)

print("Accuracy: ", accuracy_score(y_test, prediction))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

print("train size", len(X_train))
print("test size", len(X_test))

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
prediction = decision_tree.predict(X_test)
prediction_proba = decision_tree.predict_proba(X_test)

print("Accuracy: ", accuracy_score(y_test, prediction))
#print("AUC: ", roc_auc_score(y_test, prediction_proba[:, 1]))
```

### Grid Search

```python
from sklearn.model_selection import cross_validate, cross_val_predict
from sklearn.model_selection import GridSearchCV

parameters = {'min_samples_leaf':[1,5],'max_depth':[None,10]}

grid_search = GridSearchCV(RandomForestClassifier(n_estimators=20), parameters, cv=10)
grid_search.fit(ada_X_train, ada_y_train)

prediction = grid_search.predict(ada_X_test)
prediction_proba = grid_search.predict_proba(ada_X_test)
    
best_classifier = grid_search.best_estimator_

print('Best classifier:',best_classifier)
print('Accuracy:', accuracy_score(ada_y_test,prediction))
```
